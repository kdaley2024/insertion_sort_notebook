# Having fun with insertion sort 

Import libraries

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
```

Create a matrix to represent the image

```python
def createImage(nrows, ncols):
    # Create a matrix 
    matrix = np.zeros((nrows, ncols))

    for i in range(nrows):
        for j in range(ncols):
            matrix[i,j] = (i+1)

    return matrix
```

Shuffle the values in the columns of the matrix

```python
def shuffle_matrix(matrix):
    # Flatten the matrix
    flattened_matrix = matrix.flatten()

    # Shuffle the flattened matrix
    np.random.shuffle(flattened_matrix)

    # Reshape the matrix back to its original shape
    shuffled_matrix = flattened_matrix.reshape(matrix.shape)

    return shuffled_matrix 
```

Sort the values in the columns of the matrix 

```python
def sort_columns(matrix):

    sorted_matrix = np.sort(matrix, axis=0)

    return sorted_matrix
```

Create the matrix and display

```python
# Define the matrix size
nRows = 20
nCols = nRows

# Create a matrix to represent the image
matrix = createImage(nRows, nCols)

# Display the matrix as an image
plt.imshow(matrix, cmap='jet')
plt.colorbar()  
plt.show()
```

Shuffle the values in the matrix columns

```python

# Suffle the matrix values
shuffled_matrix = shuffle_matrix(matrix)

# Display suffled matrix as an image
plt.imshow(shuffled_matrix, cmap='jet')
plt.colorbar()  
plt.show()
```

Sort columns to recover original matrix 

```python
# Sort each column to recover the original order
sorted_matrix = sort_columns(matrix)

# Display sorted matrix as an image
plt.imshow(sorted_matrix, cmap='jet')
plt.colorbar()  
plt.show()
```
